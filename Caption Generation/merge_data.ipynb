```python
import pandas as pd
import glob
import pyarrow as pa
import pyarrow.parquet as pq

files = sorted(glob.glob("data/*.parquet"))

output_file = "data/food_caption_dataset.parquet"

writer = None

for f in files:
    print("Processing:", f)

    df = pd.read_parquet(f)

    df = df[[
        "image",
        "Attributes_ViT_L_14_descriptors_text_davinci_003_full"
    ]]

    df = df.rename(columns={
        "Attributes_ViT_L_14_descriptors_text_davinci_003_full": "caption"
    })

    df = df.explode("caption")
    df = df.dropna(subset=["caption"])

    table = pa.Table.from_pandas(df)

    if writer is None:
        writer = pq.ParquetWriter(output_file, table.schema)

    writer.write_table(table)

if writer:
    writer.close()

print("✅ Parquet dataset saved successfully")
```

```python
import pyarrow.parquet as pq
import pandas as pd

parquet_file = pq.ParquetFile("data/food_caption_dataset.parquet")

seen_ids = set()
global_idx = 0

for i, batch in enumerate(parquet_file.iter_batches(batch_size=20000)):

    df = batch.to_pandas()

    # image_id generate
    df["image_id"] = range(global_idx, global_idx + len(df))
    global_idx += len(df)

    df = df[["image_id","image"]]

    df = df[~df["image_id"].isin(seen_ids)]

    seen_ids.update(df["image_id"])

    df.to_parquet(f"images_part_{i}.parquet")
```

```python
parquet_file = pq.ParquetFile("data/food_caption_dataset.parquet")
```

```python
seen_ids = set()

for i, batch in enumerate(parquet_file.iter_batches(batch_size=20000)):

    df = batch.to_pandas()

    df["image_id"] = range(i*20000, i*20000 + len(df))

    # -------- IMAGES --------
    images = df[["image_id","image"]]
    images.to_parquet(f"images_part_{i}.parquet")

    # -------- CAPTIONS --------
    caption_col = [c for c in df.columns if "descriptors_text_davinci" in c][0]

    captions = df[["image_id", caption_col]]

    captions = captions.explode(caption_col)

    captions = captions.rename(columns={caption_col:"caption"})

    captions.to_parquet(f"captions_part_{i}.parquet")

    print(f"Processed batch {i}")
```
